> ### **Download the data: gdwon package**

```python
import gdown  #gdown패키지는 구글 드라이브를 포함하여 웹에서 대용량 파일을 다운로드할 수 있는 패키지입니다.
gdown.download('https://bit.ly/3eecMKZ', '남산도서관 장서 대출목록 (2021년 04월).csv', quiet=False) #quiet=False 다운로드 진행상황 출력
```

> ### **Check file encoding format:chardet.detect() function**









```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', mode='rb') as f:
  d = f.readline()
print(chardet.detect(d))
```

**영어는 보통 UTF-8 사용** (전세계 모든 문자를 컴퓨터에 표현하기 위해 만들어짐, 최대 4바이트까지 사용)

**한국어는 EUC-KR 사용** (한글을 위한 완성형 인코딩 중 하나 2바이트 사용)

>### **Specify encoding format**

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR') as f:
  print(f.readline())
  print(f.readline())
```

> ### **Read a CSV file into a DataFrame**: read_csv()function

```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR',dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str})

df.head()
```

> ### **Save DataFrame to a CSV file**

```python
df.to_csv('ns_202104.csv')
```
